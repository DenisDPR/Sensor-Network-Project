```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline 


#Import the data
LR_Result_df = pd.read_csv('dfD_v3_CGDD_CYIELD.csv')

x = LR_Result_df['Cum_GDD']
y = LR_Result_df['Cum_Yield']
xD = LR_Result_df['date']

#Fit
fit = np.polyfit(x,y,3)
fit_fn = np.poly1d(fit)

sns.set_style("darkgrid")

#Plot
plt.plot(x,y,label='data',color='r')
plt.plot(x,fit_fn(x),color='b',label='fitted')
plt.legend(loc='upper left')
```
